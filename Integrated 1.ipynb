```python
#Convert the data to the required types
#we start with 'year_of_release' that we convert from non-null float64 to astype(np.int64)  to keep only the year
#we need first to replace the nan before attempting the conversion with  df.fillna(df.mean(), inplace=True)
game_data['year_of_release'].describe()
```

For that section 'year_of_release' we have 269 rows empty around 1.6% missing values and not much difference between mean and median .

```python
#let's check the missing values for year of release
game_data[game_data['year_of_release'].isnull()]
```

```python
#We could check if we don't have duplicate games in other platforms to see if we can replace some with the latest year
#1st  example 'Madden NFL 2004 
game_data[game_data['name']=='Madden NFL 2004']
```

We can see some games exist in other platforms, lets replace the missing values with the most recent years for the duplicates

```python
#replace the missing values with the most recent years for the duplicates
game_data['year_of_release'] = game_data['year_of_release'].replace(np.nan, game_data.groupby('name')['year_of_release'].transform('max'))
```

```python
#checking  Madden NFL 2004 
game_data[game_data['name']=='Madden NFL 2004']
```

```python
#checking LEGO Batman: The Videogame
game_data[game_data['name']=='LEGO Batman: The Videogame']
```

```python
game_data[game_data['year_of_release'].isnull()]
```

After filling up the missing values for duplicates we still have **0.87% or 146 rows with empty values** for years of release and we saw earlier there not much difference between mean and median. <br>
Therefore we could replace them with the mean **2006**.

```python
#replace remaining missing values   with the mean 2006.
game_data['year_of_release']= game_data['year_of_release'].fillna(game_data['year_of_release'].mean())
game_data['year_of_release'] =game_data['year_of_release'].astype(np.int64) 
 
```

```python
#we can see these 2 rows have missing values in few columns 
# for the names to keep a certain authenticity I will replace the missing values by unknown 
game_data['name']= game_data['name'].fillna('unknown')
#checking number of values counts in genre category to select the most popular  
game_data['genre'].value_counts()
```

```python
#replacing missing values in genre column  by Action as it is the most popular genre at 20 % of total genre
game_data['genre']= game_data['genre'].fillna('Action')
```

```python
game_data['user_score'].unique()
```

In the column user_score we have identified 2 types of missing values:
* **nan**
* **tbd** (to be determined )<br>
We need to replace tbd by nan to get a numeric format so we can analyse them

```python
# replace tbd by nan
game_data['user_score']= game_data['user_score'].replace('tbd',np.NaN)
```

```python
#checking if there are still some tbd
game_data['user_score'].unique()
```

For filling the missing values in **user_score** we are going to use **Linear interpolation** which makes a mean between the values before the missing data and the value after.<br>
In fact we are going to guess the value that is most likely to be filled in.<br>
This way we are going to extrapolate the missing data in the absence of context to help us to infer the relationship between the values.<br>

```python
game_data['user_score']=game_data['user_score'].astype(float).interpolate(method='linear', direction = 'forward') 
```

```python
game_data['user_score'].unique()
```

```python
#rounding to 1 decimal for simplicity
game_data['user_score']=game_data['user_score'].round(1)
```

### **<u>Conclusion:</u>**<br>
We have prepared the data for further analysis by lowercasing column names via map function.<br>
Then we converted the data type of few columns:
* **'year_of_release'**: from float64 to astype(np.int64) to keep the year without the decimal 
* **'critic_score'**:  from non-null float64 to astype(np.int64)  to avoid decimal as well 
After that we dealt with the missing values for each column:
* **'year_of_release'** : we proceeded in 2 phases, first we looked for duplicates with missing years and replaced them with the most recent year of the duplicate. Then we filled in the remaining nan with the mean value as there was not much missing value left only 0.8%
* **'critic_score'**: we chose to use as well the mean as there was not much difference between the mean and median 
* **'name'**: we chose to use unknown for the missing values as there were only 2 rows missing and we wanted to keep them
* **'genre'**: we chose to categorize them into  Action as it is the most popular genre at 20 % of total 
* **'user_score'**: we replaced first'tbd' by Nan then all the missing were filled in through the Linear interpolation to keep a certain coherence in the relationship between values when there were so many missing rows  
* **'rating'** : we didn't fill the missing values instead we created another parameter within the column called N/A which, contain all missing values in this section

We identified a quite significant number of missing values in certain sections which has made the imputation difficult sometimes. <br>
We didn't want to just delete those because of the fear that the dataset can lose expressiveness and lead to weak or biased analyses.<br>
In this dataset the causes of  missing values could be the result of a faulty extraction or import of the data, or it could be the result of the collection process. <br>Maybe many surveys might have been left out blank by respondents.<br>
From a quick data profiling it seems that the potential reasons why we have missing values could be related to the year of release , the older a game is the less data we were able to gather( example: year 1980) or simply there were simply not enough data available at the time of the survey for that particular game. ###  **<u>Formulate hypothese : Case 1: Average user ratings of the Xbox One and PC platforms are the same.</u>**

research question : Does the average user ratings  from  Xbox One  and PC are the same?

* If the p-value is less than 0.05, we reject the null hypothesis that there's no difference between the means and conclude that a significant difference does exist.

* If the p-value is larger than 0.05, we cannot conclude that a significant difference exists.

population => n ratings of Xbox One and ratings of PC<br>
parameter of interest => user ratings<br>
(H0)null hypothesis => (H0) = μ1= μ2 The null hypothesis (H0) for the hypothese is that the proportions are the same.<br>
(H1)alternative hypothesis => (H1)= μ1 ≠ μ2<br>

p-value<α ⇒ reject H0 <br>
p-value≥α ⇒ fail to reject H0<br>

parameter of interest => users ratings  from  Xbox One, users ratings  from  PC<br>
(H0)null hypothesis => The null hypothesis (H0) for the test is that both average ratings are the same .<br>
(H1)alternative hypothesis => The alternative hypothesis is that (H1)= rating from  Xbox One differ from the one of PC

```python
#we extract the users and mean of each platform 
#PC 
pc_users_rating = game_data[game_data["platform"] == 'PC'].dropna()['user_score'].values

print(f"number of PC ratings={len(pc_users_rating )}") 
print(f"pc mean={pc_users_rating.mean()}") 
```

```python
#Xbox one
Xbox_One_users_rating = game_data[game_data["platform"] == 'XOne'].dropna()['user_score'].values
print(f"number of Xbox one ratings={len(Xbox_One_users_rating )}") 
print(f"Xbox one mean={Xbox_One_users_rating.mean()}") 
```

```python
# Testing a hypothesis about the  pc mean and Xbox one mean based on samples taken from them
# pc_users_rating is the sample from the first statistical population
# Xbox_One_users_rating is the sample from the second statistical population
# equal_var defines whether or not the variances of the samples are considered equal; default value is True 
# alpha = 0.05 ,an alpha of 0.05 is used as the cutoff for significance

from scipy import stats as st
Xbox_One_users_rating
pc_users_rating
alpha = 0.05   

results = st.ttest_ind(
    Xbox_One_users_rating,
    pc_users_rating,
    equal_var=True) 


print('p-value:',results.pvalue)
if results.pvalue < alpha:
    print('we reject the null hypothesis (H0),therefore the averages of ratings for both consoles are different')
    print(f"pc mean={pc_users_rating.mean()}")
    print(f"Xbox one mean={Xbox_One_users_rating.mean()}") 
else:
      print('we failed to reject the null hypothesis (H0) that the averages of ratings for both consoles are the same,\n but it doesnt mean we accept the alternative hypothesis (H1),\n as P-value> 0.05: it just indicates weak evidence against the null hypothesis')

```

####  <u>**Conclusion case 1**</u>
For this first hypothese on the Equality of the Means of Two Populations we applied the method spicy.stats.ttest_ind().<br>
We assumed that the samples were taken from populations with similar parameters that's why we set up equal_var= True. <br>we have specified indeed that the variances of the population should be considered also as equal.<br>
We used the p-value to decide whether or not to reject the null hypothesis because it gives us a probability of getting the desired outcome that the null hypothesis H0 is true or a different outcome further from the expectations.<br>
Finally we estimated that 5% was sufficient as conventionnal threshold in our analysis that's why we specified  alpha = 0.05.<br>

Because of the result p-value: 0.00014821566841763265 we rejected the null hypothesis (H0),therefore the averages of ratings for both consoles are different.
